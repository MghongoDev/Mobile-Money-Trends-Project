# Mobile Money Adoption Exploratory Analysis

This notebook provides an interactive exploration of mobile money adoption trends.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path

from mobile_money_project.etl import run_mobile_money_etl
from mobile_money_project.modeling import compare_models, explain_model

# Load data
df, summary = run_mobile_money_etl()
print(f"Loaded {len(df)} records")
print("Summary:", summary)
```

```python
# Data exploration
df.head()
```

```python
# Visualize trends
if 'year' in df.columns:
    plt.figure(figsize=(12, 6))
    if 'mobile_money_share' in df.columns:
        sns.lineplot(data=df, x='year', y='mobile_money_share')
    plt.title('Mobile Money Adoption Trends')
    plt.show()
```

```python
# Model comparison
model_results = compare_models(df)
print("Model Comparison:")
for model, metrics in model_results.items():
    print(f"{model}: MAE={metrics['mae']:.3f}, R²={metrics['r2']:.3f}")
```
